```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
count, ax = plt.subplots()
ax = sns.countplot(tips, x='day')
ax.set_title("Count of days")
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
sns.lmplot(tips, x='total_bill', y='tip', fit_reg=False)
```

```python
sns.lmplot(tips, x='total_bill', y='tip')
```

```python
sns.lmplot(tips, x='total_bill', y='tip', hue='sex', fit_reg=False)
```

```python
joint = sns.jointplot(tips, x='total_bill', y='tip')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle("Joint Plot of Total Bill and Trip", fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(tips, x='total_bill', y='tip', kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle("Hexbin Joint Plot of Total Bill and Trip", fontsize=10, y=1.03)
```

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(tips, x='total_bill', y='tip', fill=True)
ax.set_title("Kernel Density Plot of Total Bill and Tip")
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
kde_joint = sns.jointplot(tips, x='total_bill', y='tip', kind='kde')
```

```python
bar, ax = plt.subplots()
ax = sns.barplot(tips, x='time', y='total_bill')
ax.set_title("Bar Plot of average total bill for time of day")
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
box, ax = plt.subplots()
ax = sns.boxplot(tips, x='time', y='total_bill')
ax.set_title("Boxplot of total bill by time of day")
ax.set_xlabel('Time of day')
ax.set_ylabel('Total bill')
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.histplot)
```

```python
violin, ax = plt.subplots()
ax = sns.violinplot(tips, x='time', y='total_bill', hue='sex', split=True)
```

```python
scatter = sns.lmplot(tips, x='total_bill', y='tip', hue='sex', fit_reg=False, markers=['o', 'x'])
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(tips, x='time', y='total_bill', hue='sex', split=True)
```

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(tips, x='time', y='total_bill', ax=ax)
        violin.set_title(style)
fig.tight_layout()
```
